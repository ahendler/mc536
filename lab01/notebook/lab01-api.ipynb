### Lichess
Lichess.com é um site de xadrez online. Abaixo é mostrada uma requisição de informações sobre um usuário que é disponibilizada no formato JSON.

```python
import requests
import json
url = "https://lichess.org/api/user/hendler_a"
data = requests.get(url).content
parsed = json.loads(data)
print(json.dumps(parsed, indent=4, sort_keys=True))
```

### Road Safety Data 
Informações sobre acidentes de carro na Inglaterra em 2018 no formato CSV.

```python
import requests
import pandas as pd
import io
url = "http://data.dft.gov.uk/road-accidents-safety-data/dftRoadSafetyData_Casualties_2018.csv"
data = requests.get(url).content.decode('utf-8')
data_csv = io.StringIO(data)
data = pd.read_csv(data_csv)
print(data)
```
